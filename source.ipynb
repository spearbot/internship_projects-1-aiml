```python
class tumor_classifier(torch.nn.Module):

    def __init__(self,num_inputs,num_outputs):
        super(tumor_classifier,self).__init__()

        self.first_hidden_layer = torch.nn.Linear(in_features=num_inputs,out_features=15)
        self.first_hidden_layer_activation = torch.nn.ReLU()
        self.output_layer = torch.nn.Linear(in_features=15,out_features=1)
        self.output_layer_activation = torch.nn.Sigmoid()

    def forward(self,x):

        first_hidden_layer_out = self.first_hidden_layer(x)
        first_hidden_activation_out = self.first_hidden_layer_activation(first_hidden_layer_out)
        output_layer_out = self.output_layer(first_hidden_activation_out)
        y_hat= self.output_layer_activation(output_layer_out)

        return y_hat
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
our_nn = tumor_classifier(num_inputs=15, num_outputs=1).to(device)
```

```python
print(summary(our_nn,input_size=(15,)))
```

```python
X_train = torch.tensor(X_train,dtype=torch.float64).to(device)
Y_train = torch.tensor(Y_train,dtype=torch.float64).to(device)
X_test = torch.tensor(X_test,dtype=torch.float64).to(device)
Y_test = torch.tensor(Y_test,dtype=torch.float64).to(device)
```

```python
loss_function = torch.nn.BCELoss()
optimizer = torch.optim.SGD(params=our_nn.parameters(), lr=0.01)
epochs = 50
mb_size = X_train.shape[0]

for i in range(epochs):

    Y_hat = our_nn(X_train)
    train_loss = loss_function(Y_hat,Y_train)

    Y_hat_test = our_nn(X_test)
    test_loss = loss_function(Y_hat_test,Y_test)

    train_loss.backward()
    optimizer.step()
    optimizer.zero_grad()

    print("Epoch # {}, Train Loss Value = {}, Test Loss Value = {}".format(i+1,train_loss,test_loss))
```
